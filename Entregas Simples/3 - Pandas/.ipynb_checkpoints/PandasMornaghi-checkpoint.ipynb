Construir un objeto de series de tiempo con el índice igual al año de los juegos olímpicos.

```python
import pandas as pd
```

```python
df = pd.read_csv("Summer-Olympic-medals-1976-to-2008.csv")
df
```

```python
year_datetime = pd.to_datetime(df['Year'], format = '%Y')
year_datetime
```

```python
Serie_Olympics = pd.Series(df['City'].values, index = year_datetime)
Serie_Olympics
```

Separar nombres de apellidos en dos columnas distintas usando la coma como separador.

```python
namesplit = df.Athlete.str.split(",",1,expand=True)
namesplit
```

```python
df.insert(6, "First Name", namesplit[1])
df.insert(7, "Last Name", namesplit[0])
del df["Athlete"]
df
```

Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?

```python
df_aux = df.set_index('Year')
df_aux.describe(include='all')
```

```python
df['Country_Code'].value_counts(dropna=True).idxmax()
```

Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.

```python
df_mmedal = df[df['Gender']=='Men']
df_mmedal
```

```python
medalsy = df_mmedal.groupby('Year')['Medal'].count()
df_medalsy = pd.DataFrame(medalsy)
df_medalsy.columns = ['Medals by men']
df_medalsy
```
