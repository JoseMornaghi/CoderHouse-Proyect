```python
import numpy as np
import pandas as pd
```

```python
wdf = pd.read_csv('w_data.csv', sep = ';')
wdf
```

```python
pdf = pd.read_csv('p_data.csv', sep = ';')
pdf
```

```python
mdf = pd.merge(pdf,wdf, on ='ID', how = 'inner')
mdf
```

1. Cuantos ID tienen en común. Nombrar cuales no tienen en común

```python
incommon = mdf.shape[0]
incommon
```

```python
mouterdf = pd.concat([pdf,wdf]).drop_duplicates(['ID'], keep = False)
mouterdf
```

2. Corregir los target a solo dos valores distintos ‘<=50k’ y ‘>50k’

```python
mdf['Target'].replace('<=50K.', '<=50K', inplace=True, regex=True)
mdf['Target'].replace('>50K.', '>50K', inplace=True, regex=True)
mdf
```

3. Calcular la proporción que hay entre personas con salarios >50k respecto del total para entre personas de distinto sexo, Raza y Educacion. Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación

```python
categorical_mdf = mdf.select_dtypes(exclude=[np.number])
categorical_mdf.describe(include='all')
```

```python
total = mdf['Target'].count()
total
```

```python
salarybysex = mdf.groupby(['Sex','Target']).count()
salarybysex
```

```python
salarybysex.drop(salarybysex.columns[1:13], axis=1, inplace = True)
salarybysex.columns = ['Number of People']
salarybysex
```

```python
salarybysex.loc['Total']= salarybysex.sum()
salarybysex
```

```python
salarybysex/total
```

```python
salarybyrace = mdf.groupby(['Race','Target','Sex']).count()
salarybyrace.drop(salarybyrace.columns[1:13], axis=1, inplace = True)
salarybyrace.columns = ['Number of People']
salarybyrace.loc['Total']= salarybyrace.sum()
salarybyrace
```

```python
salarybyrace/total
```

```python
salarybyeducation = mdf.groupby(['Education','Target','Sex']).count()
salarybyeducation.drop(salarybyeducation.columns[1:13], axis=1, inplace = True)
salarybyeducation.columns = ['Number of People']
salarybyeducation.loc['Total']= salarybyeducation.sum()
pd.set_option("display.max_rows", None, "display.max_columns", None)
salarybyeducation
```

```python
salarybyeducation/total
```
