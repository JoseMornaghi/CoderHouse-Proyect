### Leer los set de datos w_data.csv y p_data.csv, fusionarlos a partir de la variable que tengan en común (ID)

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('p_data.csv',sep=';')
df1
```

```python
df2 = pd.read_csv('w_data.csv',sep=';')
df2
```

```python
dfm = pd.merge(df1,df2, on ='ID', how='inner')
dfm
```

### 2. Corregir los target a solo dos valores distintos ‘<=50k’ y ‘>50k’

```python
dfm['Target'].unique()
```

```python
dfm['Target'].replace(' <=50K.', '<=50K', inplace=True, regex=True)
dfm['Target'].replace(' <=50K', '<=50K', inplace=True, regex=True)
dfm['Target'].unique()
```

```python
dfm['Target'].replace(' >50K.', '>50K', inplace=True, regex=True)
dfm['Target'].replace(' >50K', '>50K', inplace=True, regex=True)
dfm['Target'].unique()
```

### 3. Calcular la proporción que hay entre personas con salarios >50k respecto del total para Entre personas de distinto sexo, Raza y Educacion. Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación.


```python
sdfm = dfm[dfm['Target']=='>50K']
sdfm
```

```python
total = dfm['Target'].count()
total
```

```python
sexComparison = sdfm.groupby('Sex').count()['Target']
sexComparison
```

```python
sexResults = sexComparison / total
sexResults
```

```python
raceComparison = sdfm.groupby('Race').count()['Target']
raceComparison
```

```python
raceResults = raceComparison / total
raceResults
```

```python
educationComparison = sdfm.groupby('Education').count()['Target']
educationComparison
```

```python
educationResults = educationComparison / total
educationResults
```
