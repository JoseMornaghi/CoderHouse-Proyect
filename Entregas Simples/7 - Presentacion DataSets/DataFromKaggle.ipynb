```python
import requests 
import os
import zipfile
import pandas as pd

def download_url(url, save_path, chunk_size=128):
    r = requests.get(url, stream=True)
    with open(save_path, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=chunk_size):
            fd.write(chunk)

download_url('https://www.soa.org/globalassets/assets/files/edu/2020/2020-12-7-exam-pa-data.zip', os.path.join(os.getcwd(), "tmp.zip"))

with zipfile.ZipFile(os.path.join(os.getcwd(), "tmp.zip"), 'r') as zip_ref:
    zip_ref.extractall(os.getcwd())

dataframe = pd.read_csv('December 7 data.csv')

os.remove('December 7 data.csv')
os.remove('tmp.zip')

dataframe
```
