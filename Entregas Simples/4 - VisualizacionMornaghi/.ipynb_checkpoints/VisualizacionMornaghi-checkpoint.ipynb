# 2. Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
df_hist = df["Salary"]
df_hist
```

```python
df_flatten = df_hist.values.flatten()
df_flatten
```

```python
fig = plt.figure(figsize=(17,6))
n, bins, patches = plt.hist(df_hist, bins = 12, rwidth = 0.5)
plt.xticks(bins)
plt.title('Histograma de salarios', loc = 'left', fontsize = 22)
plt.xlabel('\nSalarios', fontsize = 14)
plt.ylabel('Cantidad de Empleados', fontsize = 14)
plt.grid()
```

Intento ordenar un poco mejor el grafico, agregar las cantidades de empleados que cobran dichos sueldos.

```python
def show_values_on_bars(axs):
    def _show_on_single_plot(ax):
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height()
            value = '{:.2f}'.format(p.get_height())
            ax.text(_x,_y,value, ha="center")
    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
        
fig, ax = plt.subplots(figsize=(17,6))
ax.hist(x=df["Salary"], bins = 12, rwidth = 0.5)
plt.title('Histograma de Salarios')
plt.xlabel('\nSalarios')
plt.ylabel('Cantidad de Empleados')
show_values_on_bars(ax)
plt.grid()
```

# 4. Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)

```python
# Agrupo los datos de Date of Hire y en base al EmpID, cuento la cantidad de empleados que se contrataron en cada fecha.
dfGroup = df.groupby('DateofHire')['EmpID'].count()
dfGroup
```

```python
df_DoH = pd.DataFrame(dfGroup)
df_DoH.columns = ['Employees by DateofHire']
df_DoH
```

```python
year_datetime = pd.to_datetime(df_DoH.index, format='%m/%d/%Y')
year_datetime
```

```python
Serie_hire = pd.Series(df_DoH['Employees by DateofHire'].values,index=year_datetime)
Serie_hire_sorted = Serie_hire.sort_index()
Serie_hire_sorted
```

```python
fig, ax = plt.subplots(figsize=(18, 6))
ax.plot(Serie_hire_sorted.index , Serie_hire_sorted.values, label='Employees Hired')
ax.set_xlabel('Year')
ax.set_ylabel('Employees Hired')
ax.set_title('Employees Hired over time')
ax.legend()
plt.grid()
```

Intento plotear con un histograma

```python
fecha_contratacion = pd.to_datetime(df['DateofHire'], format='%m/%d/%Y')
fecha_contratacion
```

```python
array_tiempo = fecha_contratacion.values.flatten()
fig,ax = plt.subplots(figsize=(18,6))
ax.hist(array_tiempo, bins = 12, rwidth = 0.5)
ax.set_xlabel('Year')
ax.set_ylabel('Employees Hired')
ax.set_title('Employees Hired over time')
plt.grid()
```

```python
def show_values_on_bars(axs):
    def _show_on_single_plot(ax):
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height()
            value = '{:.2f}'.format(p.get_height())
            ax.text(_x,_y,value, ha="center")
    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
        
array_tiempo = fecha_contratacion.values.flatten()
fig,ax = plt.subplots(figsize=(18,6))
ax.hist(array_tiempo, bins = 12, rwidth = 0.5)
ax.set_xlabel('Year')
ax.set_ylabel('Employees Hired')
ax.set_title('Employees Hired over time')
show_values_on_bars(ax)
plt.grid()
```
